```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Configuração do estilo dos gráficos
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Carregar os dados
df = pd.read_csv('iiot_30min_norm.csv', sep=';', parse_dates=['TIME'])

print("=== ANÁLISE EXPLORATÓRIA INICIAL ===")
print(f"Shape do dataset: {df.shape}")
print("\nPrimeiras 5 linhas:")
print(df.head())
print("\nÚltimas 5 linhas:")
print(df.tail())

print("\n=== ESTATÍSTICAS DESCRITIVAS ===")
print(df.describe())

print("\n=== INFO DO DATAFRAME ===")
print(df.info())

# Verificar valores nulos
print("\n=== VALORES NULOS ===")
print(df.isnull().sum())

# Configurar TIME como índice
df.set_index('TIME', inplace=True)

# Remover coluna EPOCH para análise
df_analysis = df.drop('EPOCH', axis=1)

# Função para criar cartas de controle
def criar_carta_controle(serie, titulo, L=3):
    """
    Cria cartas de controle (X-bar e S) para uma série temporal
    """
    # Calcular estatísticas
    media = serie.mean()
    std = serie.std()
    n = len(serie)

    # Limites de controle
    UCL = media + L * std
    LCL = media - L * std
    CL = media

    # Identificar pontos fora de controle
    pontos_fora_controle = serie[(serie > UCL) | (serie < LCL)]

    # Criar figura
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10))

    # Gráfico 1: Série temporal com limites
    ax1.plot(serie.index, serie.values, 'b-', linewidth=1, label='Valores')
    ax1.axhline(y=UCL, color='r', linestyle='--', label=f'UCL: {UCL:.3f}')
    ax1.axhline(y=CL, color='g', linestyle='-', label=f'CL: {CL:.3f}')
    ax1.axhline(y=LCL, color='r', linestyle='--', label=f'LCL: {LCL:.3f}')

    # Destacar pontos fora de controle
    if not pontos_fora_controle.empty:
        ax1.scatter(pontos_fora_controle.index, pontos_fora_controle.values,
                   color='red', s=50, zorder=5, label='Fora de Controle')

    ax1.set_title(f'Carta de Controle - {titulo}', fontsize=14, fontweight='bold')
    ax1.set_ylabel('Valor Normalizado')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Gráfico 2: Histograma com distribuição
    ax2.hist(serie.values, bins=30, alpha=0.7, density=True, color='skyblue')
    ax2.axvline(media, color='g', linestyle='-', label=f'Média: {media:.3f}')
    ax2.axvline(media + std, color='orange', linestyle='--', label=f'±1σ: {std:.3f}')
    ax2.axvline(media - std, color='orange', linestyle='--')
    ax2.set_xlabel('Valor Normalizado')
    ax2.set_ylabel('Densidade')
    ax2.set_title(f'Distribuição - {titulo}')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Estatísticas
    print(f"\n=== ESTATÍSTICAS PARA {titulo} ===")
    print(f"Média (CL): {media:.4f}")
    print(f"Desvio Padrão: {std:.4f}")
    print(f"UCL (+{L}σ): {UCL:.4f}")
    print(f"LCL (-{L}σ): {LCL:.4f}")
    print(f"Coeficiente de Variação: {(std/media)*100:.2f}%")
    print(f"Pontos fora de controle: {len(pontos_fora_controle)}/{n} ({(len(pontos_fora_controle)/n)*100:.1f}%)")

    if not pontos_fora_controle.empty:
        print("\nPontos fora de controle:")
        for idx, valor in pontos_fora_controle.items():
            print(f"  {idx}: {valor:.4f}")

    return {
        'media': media,
        'std': std,
        'UCL': UCL,
        'LCL': LCL,
        'pontos_fora': pontos_fora_controle
    }

# Análise de correlação
print("\n=== MATRIZ DE CORRELAÇÃO ===")
correlation_matrix = df_analysis.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, fmt='.3f')
plt.title('Matriz de Correlação entre Variáveis', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Análise de tendência temporal
print("\n=== ANÁLISE DE TENDÊNCIA TEMPORAL ===")
fig, axes = plt.subplots(4, 2, figsize=(15, 12))
axes = axes.flatten()

for i, coluna in enumerate(df_analysis.columns):
    if i < len(axes):
        axes[i].plot(df_analysis.index, df_analysis[coluna], linewidth=1)
        axes[i].set_title(f'Evolução Temporal - {coluna}')
        axes[i].set_ylabel('Valor Normalizado')
        axes[i].tick_params(axis='x', rotation=45)
        axes[i].grid(True, alpha=0.3)

# Remove o último subplot vazio se necessário
if len(df_analysis.columns) < len(axes):
    for i in range(len(df_analysis.columns), len(axes)):
        fig.delaxes(axes[i])

plt.tight_layout()
plt.show()

# Cartas de controle para cada variável
print("\n" + "="*60)
print("CARTAS DE CONTROLE INDIVIDUAIS")
print("="*60)

resultados_controle = {}

for coluna in df_analysis.columns:
    print(f"\n{'='*50}")
    print(f"ANALISANDO: {coluna}")
    print(f"{'='*50}")

    resultados_controle[coluna] = criar_carta_controle(df_analysis[coluna], coluna)

# Análise resumida
print("\n" + "="*60)
print("RELATÓRIO RESUMIDO DE CONTROLE ESTATÍSTICO")
print("="*60)

resumo_df = pd.DataFrame({
    'Variável': df_analysis.columns,
    'Média': [resultados_controle[col]['media'] for col in df_analysis.columns],
    'Desvio_Padrão': [resultados_controle[col]['std'] for col in df_analysis.columns],
    'UCL': [resultados_controle[col]['UCL'] for col in df_analysis.columns],
    'LCL': [resultados_controle[col]['LCL'] for col in df_analysis.columns],
    'Pontos_Fora_Controle': [len(resultados_controle[col]['pontos_fora']) for col in df_analysis.columns],
    '%_Fora_Controle': [(len(resultados_controle[col]['pontos_fora'])/len(df_analysis))*100 for col in df_analysis.columns]
})

print(resumo_df.round(4))

# Identificar variáveis com maior instabilidade
print("\n=== VARIÁVEIS COM MAIOR INSTABILIDADE ===")
instabilidade = resumo_df.nlargest(3, 'Desvio_Padrão')[['Variável', 'Desvio_Padrão', '%_Fora_Controle']]
print(instabilidade)

# Análise de capacidade do processo (considerando limites 0-1)
print("\n=== ANÁLISE DE CAPACIDADE DO PROCESSO ===")
for coluna in df_analysis.columns:
    dados = df_analysis[coluna]
    # Considerando que os dados são normalizados entre 0-1
    LSE = 1.0  # Limite Superior de Especificação
    LIE = 0.0  # Limite Inferior de Especificação

    cp = (LSE - LIE) / (6 * dados.std())
    cpk = min((LSE - dados.mean()) / (3 * dados.std()),
              (dados.mean() - LIE) / (3 * dados.std()))

    print(f"\n{coluna}:")
    print(f"  Cp: {cp:.3f}")
    print(f"  Cpk: {cpk:.3f}")
    if cpk >= 1.33:
        print("  ✅ Processo Capaz")
    elif cpk >= 1.0:
        print("  ⚠️  Processo Marginalmente Capaz")
    else:
        print("  ❌ Processo Não Capaz")

# Gráfico resumo dos limites de controle
fig, ax = plt.subplots(figsize=(12, 6))
variáveis = resumo_df['Variável']
ucl_values = resumo_df['UCL']
lcl_values = resumo_df['LCL']
media_values = resumo_df['Média']

x_pos = np.arange(len(variáveis))
width = 0.25

ax.bar(x_pos - width, media_values, width, label='Média (CL)', alpha=0.8)
ax.bar(x_pos, ucl_values, width, label='UCL', alpha=0.6)
ax.bar(x_pos + width, lcl_values, width, label='LCL', alpha=0.6)

ax.set_xlabel('Variáveis')
ax.set_ylabel('Valores')
ax.set_title('Comparação dos Limites de Controle por Variável', fontweight='bold')
ax.set_xticks(x_pos)
ax.set_xticklabels(variáveis, rotation=45)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("ANÁLISE CONCLUÍDA")
print("="*60)
```
